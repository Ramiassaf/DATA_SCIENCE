# Importing The Needed Libraries

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
import plotly.colors as colors
pio.templates.default = "plotly_white"
```

# Loading The Dataset

```python
df = pd.read_csv("/content/drive/MyDrive/DATA/Sample - Superstore.csv", encoding="latin_1")
```

 **Utilizing the 'Order Date' column in the dataset provides a valuable opportunity to enhance our sales and profit analysis. By extracting key temporal information, such as the order month, order year, and order day, we can gain deeper insights into how sales and profits vary over time. These additional columns will prove invaluable in our data-driven decision-making process.**

**Let's proceed to enrich our dataset by incorporating these temporal dimensions. This enhancement will empower us to conduct more granular and insightful sales and profit analyses, facilitating better-informed strategies and driving business growth.**

```python
from pandas._libs.tslibs.offsets import MonthBegin
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Ship Date"] = pd.to_datetime(df["Ship Date"])

df["Order Month"] = df["Order Date"].dt.month # dt is used to access data time component
df["Order Year"] = df["Order Date"].dt.year
df["Order Day of Week"]= df["Order Date"].dt.dayofweek
```

## Monthly Sales

```python
sales_bymonth = df.groupby("Order Month")["Sales"].sum().reset_index()
fig = px.line(sales_bymonth, x="Order Month", y="Sales", title="Monthly Sales Analysis", color_discrete_sequence=["#458B74"])
fig.show()
```

```python
sales_by_category = df.groupby('Category')['Sales'].sum().reset_index()


fig = px.pie(sales_by_category,
             values='Sales',
             names='Category',
             hole=0.5,
             color_discrete_sequence=px.colors.qualitative.Pastel)

fig.update_traces(textposition='inside', textinfo='percent+label')
fig.update_layout(title_text='Sales Analysis by Category', title_font=dict(size=24))

fig.show()
```

```python
sales_by_subcategory = df.groupby('Sub-Category')['Sales'].sum().reset_index()
sales_by_subcategory = sales_by_subcategory.sort_values(by='Sales')
fig = px.bar(sales_by_subcategory,
             x='Sub-Category',
             y='Sales',
             title='Sales Analysis by Sub-Category',
              color_discrete_sequence=["#458B74"])
fig.show()
```

## Monthly Profit

```python
profit_month = df.groupby("Order Month")["Profit"].sum().reset_index()
fig = px.line(profit_month,
              x= "Order Month",
              y= "Profit",
              title = "Monthly Profit Analysis",
              color_discrete_sequence = ["#00FFFF"])
fig.show()
```

```python
profit_by_category = df.groupby('Category')['Profit'].sum().reset_index()

fig = px.pie(profit_by_category,
             values='Profit',
             names='Category',
             hole=0.5,
             color_discrete_sequence=px.colors.qualitative.Pastel)

fig.update_traces(textposition='inside', textinfo='percent+label')
fig.update_layout(title_text='Profit Analysis by Category', title_font=dict(size=24))

fig.show()
```

```python
profit_by_subcategory = df.groupby('Sub-Category')['Profit'].sum().reset_index()
profit_by_subcategory = profit_by_subcategory.sort_values(by='Profit')
fig = px.bar(profit_by_subcategory, x='Sub-Category',
             y='Profit',
             title='Profit Analysis by Sub-Category',
             color_discrete_sequence=["#00FFFF"])
# fig.update_xaxes(tickangle=30)  # Set the angle of rotation for x-axis labels

fig.show()
```

```python
sales_profit_by_segment = df.groupby('Segment').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()

color_palette = colors.qualitative.Pastel

fig = go.Figure()
fig.add_trace(go.Bar(x=sales_profit_by_segment['Segment'],
                     y=sales_profit_by_segment['Sales'],
                     name='Sales',
                     marker_color=color_palette[8]))
fig.add_trace(go.Bar(x=sales_profit_by_segment['Segment'],
                     y=sales_profit_by_segment['Profit'],
                     name='Profit',
                     marker_color=color_palette[10]))

fig.update_layout(title='Sales and Profit Analysis by Customer Segment',
                  xaxis_title='Customer Segment', yaxis_title='Amount')

fig.show()
```

**So the store has higher profits from the product sales for consumers, but the profit from corporate product sales is better in the sales-to-profit ratio. Let’s have a look at it to validate our findings:**

```python
sales_profit_by_segment = df.groupby('Segment').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()
sales_profit_by_segment['Sales_to_Profit_Ratio'] = sales_profit_by_segment['Sales'] / sales_profit_by_segment['Profit']
print(sales_profit_by_segment[['Segment', 'Sales_to_Profit_Ratio']])
```
